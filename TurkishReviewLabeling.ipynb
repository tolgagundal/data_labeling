```python
import pandas as pd

# Step 1: I create a small dataset of 15 Turkish customer reviews.
# The goal is to simulate a manual data labeling process for sentiment analysis.

reviews = [
    "Teslimat çok hızlıydı, teşekkürler!",  # positive
    "Paket geç geldi, memnun kalmadım.",    # negative
    "Ürün iyi ama paketlemesi özensizdi.",  # neutral
    "Mükemmel hizmet!",                     # positive
    "Ne iyi ne kötüydü.",                   # neutral
    "Kargo kayboldu, ulaşamadım.",          # negative
    "Süper hızlı geldi, çok memnun kaldım!",# positive
    "Yemek soğuktu ve geç geldi.",          # negative
    "Kurye çok nazikti.",                   # positive
    "İade süreci biraz uğraştırıcıydı.",    # neutral
    "Beklediğimden daha iyiydi!",           # positive
    "Tatlı güzel ama porsiyon küçüktü.",    # neutral
    "Hiç memnun kalmadım, son siparişimdi.",# negative
    "Her şey yolundaydı, teşekkür ederim.", # positive
    "Siparişim eksik geldi."                # negative
]
```

```python
# Step 2: I create a DataFrame and initialize an empty 'label' column
df = pd.DataFrame({'review': reviews})
df['label'] = ""  # We will manually label each review
```

```python
# Step 3: I assign manual sentiment labels
labels = [
    "positive", "negative", "neutral", "positive", "neutral",
    "negative", "positive", "negative", "positive", "neutral",
    "positive", "neutral", "negative", "positive", "negative"
]

df['label'] = labels
```

```python
# Step 4: Show the labeled dataset
df.reset_index(drop=True, inplace=True)
df.head(15)
```
